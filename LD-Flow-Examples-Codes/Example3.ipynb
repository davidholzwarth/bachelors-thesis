```python
# ---- iterative power propagation

p_A, p_B, p_C = 0, 0, 0

num_iterations = 10

for i in range (num_iterations):
    # each iteration, add one more power to each person, and do the delegations
    p_A = p_A + 1
    p_B = p_B + 1 
    p_C = p_C + 1

    # snapshot of the current power distribution
    temp_A = p_A
    temp_B = p_B
    temp_C = p_C

    # delegation A->B
    p_B += temp_A
    p_A -= temp_A

    # delegation B-0.5->B
    p_B += 0.5 * temp_B
    p_B -= 0.5 * temp_B

    # delegation B-0.5->C
    p_C += 0.5 * temp_B
    p_B -= 0.5 * temp_B

print("1:")
print(p_A / num_iterations, 
    p_B / num_iterations, 
    p_C / num_iterations)

# ---- iterative power propagation with no source effect

p_A, p_B, p_C = 1, 1, 1

num_iterations = 10

for i in range (num_iterations):
    # snapshot of the current power distribution
    temp_A = p_A
    temp_B = p_B
    temp_C = p_C

    # delegation A->B
    p_B += temp_A
    p_A -= temp_A

    # delegation B-0.5->B
    p_B += 0.5 * temp_B
    p_B -= 0.5 * temp_B

    # delegation B-0.5->C
    p_C += 0.5 * temp_B
    p_B -= 0.5 * temp_B

print("2:")
print(p_A, p_B, p_C)


```
